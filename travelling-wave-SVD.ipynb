```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dat=pd.read_csv("/content/drive/My Drive/data/dat.txt",header=None,sep=' ')
dat=dat.drop(512, axis=1)
```

```
X=np.array(dat)
X.shape
print(np.linalg.matrix_rank(X))
```

```
x=np.linspace(-1,1,512)
t=[0,100,200,300,400,500,600]
```

```
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams.update({'font.size': 18})
for i in t:
  plt.plot(x,X[i])
plt.title('Fisher-KPP')
plt.show()
```

```
U, S, VT = np.linalg.svd(X,full_matrices=0)
```

```
plt.semilogy(S,'-o',color='k')
plt.ylim((10**(-16),10**(4)+1))
plt.yticks(np.power(10,np.arange(-16,4,4, dtype=float)))
plt.xticks(np.arange(0,1000,250))
plt.grid()
plt.title('Log singular values')
plt.show()
```

```
S=np.diag(S)
for r in [1,2,3,4,5,10,100]:
    # Construct approximate image
  Xapprox = U[:,:r] @ S[0:r,:r] @ VT[:r,:]
  for i in t:
    plt.plot(x,Xapprox[i])
    plt.title('rank = {}'.format(r))
  plt.show()
```
